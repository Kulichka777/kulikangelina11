```python
import time

score = 0
invntory = []

# Функція для виведеня введеного тексту з паузою 
def print_with_paus(text):
    print(text)
    time.sleep(1)

# Початок Гри 
print_with_paus ("Ласкаво просимо до текстовоЇ гри!" )
print_with_paus ("Ви знаходитесь у великому покинутому місті.Виберіть напрямок подорожі:")
choic = input ("1. Піти оглянути вулиці.\n2. Піти у покинуту лікарню.\n ")

if choic == '1':
    print_with_paus ("Ви відправились оглядати вулиці." )
    print_with_paus ("За декілька хвилин ви натрапили на жіночу сумочку ")
    print_with_paus ("Під час обшуку ви знайшли помаду,ключі та 10$!")
    score = 10
    invntory.append("$")
    choice = input("Хочу продовжити досліджувати вулиці.\n2.Хочу втекти з міста\n")
    if choice == '1':
        print_with_paus ("Ви продовжили оглядати вулиці ")
        print_with_paus ("По дорозі ви побачили дім з увімкненим світлом та відкритими дверми")
        print_with_paus("Увійшовши туди ви побачили 2 кімнати")
        choice = input ("1.Піти праворуч\n.2.Піти ліворуч")
        if choice == '1':
            print_with_paus("Ви потрапили в дитячу кімнату")
            print_with_paus("Оглянувшись ви побачили багато дитячих іграшок")
            print_with_paus("На одній з них була кров")
            print_with_paus("Вам стало страшно")
            print_with_paus("Чи бажаєте ви повернутися назад")
            choice = input("1. Хочу вийти\n2.Хочу оглянутись тут")
            if choice == '1':
                print_with_paus("Ви вибрали")
                print_with_paus("Ви продовжуєте свою подорож")
                print_with_paus("Гра завершена. Ваш рахунок: " + str(score))
                print_with_paus("Дякуємо за гру!")
            else:
                print_with_paus("Ви не встигли вийти двері зачинились")
                print_with_paus("Тепер ви тут назавжди")
                print_with_paus("Це кінець")

        else:
            print_with_paus("Ви не встигли вибігти вас убили")
            print_with_paus("Це кінець!")

    else:
        print_with_paus("Ви спробували втекти з міста та натрапили на злочинців, вас вбили")
        print_with_paus("Це кінець гри!")
else:
    print_with_paus("Ви вирушили у покину лікарні.")
    print_with_paus("Тут виявилось 2 шляхи")
    print_with_paus("Який шлях оберете? ")
    choice = input("1.Хочу піти на ліво.\n2.Я піду праворуч\n")

    if choice == '1':
        print_with_paus("Ви пішли ліворуч та натрпили на смерьельну пастку!")
        print_with_paus("Ви померли !")
        print_with_paus("Це кінець гри!")
    else:
        print_with_paus("Ви пішли праворуч.")
        print_with_paus("Пройшовши декілька метрів ви знайшли двері")
        print_with_paus("Увійшовши ви знайшли скарб")
        print_with_paus("Ви знайшли дорогі лікарські припарати собівартістю 1000$")
        score += 1000
        print_with_paus("Вийшовши з лікарні у вас є 2 шляхи")
        choice = input ("куди піти?\n1. Прямо в лабораторію\n2.Подорожувати далі")
        if choice == '1':
            print_with_paus("Зайшовши в лабораторію вас знайшов монстр і вбив вас")
            print_with_paus("Це кінець гри!")
        else:
            print_with_paus("Ви вийшли та продовжили свою подорож")
            print_with_paus("Ви продовжуєте свою подорож не знаючи що вас чекає.")
            print_with_paus("Гра завершена. Ваш рахунок: " + str (score))
            print_with_paus("Ваш інвентар: " + ', '.join(invntory))
            print_with_paus("Дякуємо за гру!")







        
```
